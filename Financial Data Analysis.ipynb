### Investment Multiple and CAGR

```python
#multiple = ending value /initial investment 
multiple = msft.Close[-1] / msft.Close[0]
multiple
```

```python
(multiple-1)*100
```

```python
msft.Close / msft.Close[0]
```

```python
### CAGR constant anual rate of return  
start = msft.index[0]
end = msft.index[-1]
```

```python
td = end -start
```

```python
td_years = td.days/365.25
```

```python
cagr = multiple**(1/td_years) -1
cagr
```

### Compund Returns and Geometric Mean Return

```python
multiple = (1+msft.Returns).prod()
multiple
```

```python
n =msft.Returns.count() 
n
```

```python
geo_mean = multiple**(1/n)-1
geo_mean
```
