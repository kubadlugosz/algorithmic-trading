```python
df = pd.DataFrame(columns=['Open','High','Low','Close','Volume','Complete'])
```

```python
def stream_candle(msg):
    
    event_time = pd.to_datetime(msg['E'],unit='ms')
    start_time = pd.to_datetime(msg['k']['t'],unit='ms')
    first = float(msg['k']['o'])
    high = float(msg['k']['h'])
    low = float(msg['k']['l'])
    close = float(msg['k']['c'])
    volume = float(msg['k']['v'])
    complete = float(msg['k']['x'])
    
    print("Time: {} | Price: {}".format(event_time, close))
    
    df.loc[start_time] = [first, high, low, close, volume, complete]
```

```python
twm = ThreadedWebsocketManager()
twm.start()
```

```python
twm.start_kline_socket(callback = stream_candle , symbol ='BTCUSDT',interval='1m')
```

```python
twm.stop()
```

## Creating a simple bot

```python
#import packages 
from binance.client import Client 
from binance import ThreadedWebsocketManager
import pandas as pd
import numpy as np
```

```python
client = Client(key_test,secret_test,tld='us',testnet=True)
```

```python
client.get_account()
```

```python
twm = ThreadedWebsocketManager()
twm.start()
```

```python
#we will build the bot 
def simple_bot(msg):
    
    time = pd.to_datetime(msg['E'],unit = 'ms')
    price = float(msg['c'])
    
    print('Time: {}| Price: {}'.format(time,price))
    
    if int(price) % 10 == 0:
        order = client.create_order(symbol = 'BTCUSDT', side = 'BUY', type ='MARKET',quantity = 0.1) 
        print('\n' + 50 *'-')
        print('Buy {} BTC for {} USDT'.format(order['executedQty'],order['cummulativeQuoteQty']))
        print('\n' + 50 *'-')
        
        twm.stop()
```

```python
twm.start_symbol_miniticker_socket(callback = simple_bot , symbol ='BTCUSDT')
```
